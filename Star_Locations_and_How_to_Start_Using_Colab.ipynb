````python
# Sort the data set according your constellation
myconstellation = data.query('con == "Cyg"') # I'm looking for the stars in Cygnus
plt.close('all')
fig = plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)

# Plot the stars in your constellation
plt.scatter(myconstellation['ra'],myconstellation['dec'])
plt.gca().invert_xaxis()
plt.title("A constellation in the sky")
plt.xlabel("right ascension")
plt.ylabel("declination")
````

**Learning Check!**

That is a lot of stars...why does it not look like the constellation you see in the sky?

The good news is we can sort out some of the data. Start with the 20 brightest stars. Play around with that sort function until you see your constellation.

````python
# Sort the data set according your constellation
myconstellation = data.query('con == "Cyg"') # I'm looking for the stars in Cygnus
plt.close('all')
fig = plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)

# Plot the stars in your constellation
myconstellation = myconstellation.sort_values('mag').head(6)

plt.scatter(myconstellation['ra'],myconstellation['dec'])
plt.gca().invert_xaxis()
plt.title("A constellation in the sky")
plt.xlabel("right ascension")
plt.ylabel("declination")
````

Now overlay your constellation on to the big map.



````python
#Sort by your constellation
myconstellation = data.query('con == "Cyg"') # I'm looking for the stars in Cygnus
plt.close('all')
fig = plt.figure(figsize=(30, 5))
plt.subplot(1, 2, 1)

# Plot the stars in your constellation onto the big plot
myconstellation = myconstellation.sort_values('mag').head(6)
plt.scatter(data['ra'],data['dec'], s=0.01) #Plot all of the stars

plt.scatter(myconstellation['ra'],myconstellation['dec'])
plt.title("A constellation in the sky")
plt.xlabel("right ascension")
plt.ylabel("declination")
````
